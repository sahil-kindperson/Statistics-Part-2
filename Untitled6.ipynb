```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results.

import numpy as np
from scipy.stats import norm

population_mean = 100
population_std = 15
alpha = 0.05

sample = [105, 110, 95, 100, 98, 107, 102, 108, 104, 99]
sample_mean = np.mean(sample)
sample_size = len(sample)

z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

p_value = 2 * (1 - norm.cdf(abs(z_score)))

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference between sample and population mean.")
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu_0, mu_1, sigma, n, alpha=0.05):
    """
    Visualizes Type I and Type II errors for a two-tailed Z-test.

    Parameters:
        mu_0: Population mean under null hypothesis
        mu_1: Population mean under alternative hypothesis
        sigma: Population standard deviation
        n: Sample size
        alpha: Significance level (default = 0.05)
    """
    se = sigma / np.sqrt(n)

    z_critical = norm.ppf(1 - alpha / 2)

    x_crit_low = mu_0 - z_critical * se
    x_crit_high = mu_0 + z_critical * se

    x = np.linspace(mu_0 - 4*se, mu_1 + 4*se, 1000)

    y0 = norm.pdf(x, mu_0, se)
    y1 = norm.pdf(x, mu_1, se)

    plt.figure(figsize=(12, 6))
    plt.plot(x, y0, label="H₀ Distribution", color='blue')
    plt.plot(x, y1, label="H₁ Distribution", color='green')

    plt.fill_between(x, y0, where=(x < x_crit_low) | (x > x_crit_high), color='red', alpha=0.3, label='Type I Error (α)')

    plt.fill_between(x, y1, where=(x >= x_crit_low) & (x <= x_crit_high), color='orange', alpha=0.3, label='Type II Error (β)')

    plt.fill_between(x, y1, where=(x < x_crit_low) | (x > x_crit_high), color='green', alpha=0.1, label='Power (1 - β)')

    plt.axvline(x_crit_low, color='black', linestyle='--')
    plt.axvline(x_crit_high, color='black', linestyle='--')
    plt.title("Type I and Type II Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True)
    plt.show()

    beta = norm.cdf(x_crit_high, loc=mu_1, scale=se) - norm.cdf(x_crit_low, loc=mu_1, scale=se)
    power = 1 - beta
    print(f"Type I Error (α): {alpha}")
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {power:.4f}")

    visualize_type1_type2_errors(mu_0=100, mu_1=105, sigma=15, n=30, alpha=0.05)

```

```python
#6. Write a Python program to perform an independent T-test and interpret the results.

import numpy as np
from scipy.stats import ttest_ind

np.random.seed(1)
group1 = np.random.normal(loc=100, scale=10, size=30)
group2 = np.random.normal(loc=105, scale=10, size=30)

t_stat, p_value = ttest_ind(group1, group2, equal_var=True)  # assume equal variance

print("Group 1 Mean:", np.mean(group1))
print("Group 2 Mean:", np.mean(group2))
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the two groups.")
else:
    print("Fail to reject the null hypothesis: No significant difference between the two groups.")
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy.stats import t

def confidence_interval(data, confidence=0.95):
    """
    Calculates the confidence interval for a sample mean.

    Parameters:
        data (array-like): Sample data.
        confidence (float): Confidence level (default is 0.95 for 95% CI).

    Returns:
        (mean, lower_bound, upper_bound): Tuple containing the sample mean and CI bounds.
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    sem = np.std(data, ddof=1) / np.sqrt(n)
    t_critical = t.ppf((1 + confidence) / 2, df=n-1)  # Two-tailed t critical value

    margin_of_error = t_critical * sem
    lower_bound = mean - margin_of_error
    upper_bound = mean + margin_of_error

    return mean, lower_bound, upper_bound

np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=8, size=25)

mean, lower, upper = confidence_interval(sample_data, confidence=0.95)

print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")

```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data.

import numpy as np
from scipy.stats import t

def calculate_margin_of_error(data, confidence=0.95):
    """
    Calculates the margin of error for a sample mean.

    Parameters:
        data (array-like): Sample data.
        confidence (float): Confidence level (default is 0.95 for 95%).

    Returns:
        margin_of_error (float): The margin of error.
    """
    data = np.array(data)
    n = len(data)
    sample_std = np.std(data, ddof=1)
    standard_error = sample_std / np.sqrt(n)

    t_critical = t.ppf((1 + confidence) / 2, df=n - 1)

    margin_of_error = t_critical * standard_error
    return margin_of_error


np.random.seed(10)
sample_data = np.random.normal(loc=70, scale=12, size=40)

moe = calculate_margin_of_error(sample_data, confidence=0.95)
print(f"Margin of Error (95% confidence): ±{moe:.2f}")
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayes_theorem(prior_H, likelihood_E_given_H, likelihood_E_given_not_H):
    """
    Calculate posterior probability using Bayes' Theorem.

    Parameters:
        prior_H (float): Prior probability of the hypothesis (P(H))
        likelihood_E_given_H (float): Likelihood of evidence given hypothesis (P(E|H))
        likelihood_E_given_not_H (float): Likelihood of evidence given not hypothesis (P(E|¬H))

    Returns:
        posterior (float): Posterior probability P(H|E)
    """
    prior_not_H = 1 - prior_H
    evidence = (likelihood_E_given_H * prior_H) + (likelihood_E_given_not_H * prior_not_H)

    posterior = (likelihood_E_given_H * prior_H) / evidence
    return posterior


P_D = 0.01
P_Positive_given_D = 0.99
P_Positive_given_not_D = 0.05

posterior = bayes_theorem(P_D, P_Positive_given_D, P_Positive_given_not_D)

print(f"Probability of having the disease given a positive test: {posterior:.4f}")
```

```python
#12. Perform a Chi-square test for independence between two categorical variables in Python.

import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

data = np.array([[30, 20],
                 [25, 25]])

df = pd.DataFrame(data, index=["Male", "Female"], columns=["Prefer", "Don't Prefer"])

chi2, p, dof, expected = chi2_contingency(df)

print("Contingency Table:\n", df)
print("\nExpected Frequencies:\n", pd.DataFrame(expected, index=["Male", "Female"], columns=["Prefer", "Don't Prefer"]))
print(f"\nChi-square Statistic: {chi2:.4f}")
print(f"Degrees of Freedom: {dof}")
print(f"P-value: {p:.4f}")

alpha = 0.05
if p < alpha:
    print("\nReject the null hypothesis: There is a significant association between the variables.")
else:
    print("\nFail to reject the null hypothesis: No significant association between the variables.")
```

```python
#13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.


import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed):
    """
    Calculate expected frequencies for a Chi-square test from an observed frequency table.

    Parameters:
        observed (2D array or DataFrame): Observed frequency table

    Returns:
        DataFrame: Expected frequency table
    """
    if isinstance(observed, np.ndarray):
        observed = pd.DataFrame(observed)

    row_totals = observed.sum(axis=1).values.reshape(-1, 1)
    col_totals = observed.sum(axis=0).values.reshape(1, -1)
    grand_total = observed.values.sum()

    expected = (row_totals @ col_totals) / grand_total
    expected_df = pd.DataFrame(expected, index=observed.index, columns=observed.columns)

    return expected_df

observed_data = pd.DataFrame(
    [[30, 20],
     [25, 25]],
    index=['Male', 'Female'],
    columns=['Prefer', "Don't Prefer"]
)

expected = calculate_expected_frequencies(observed_data)

print("Observed Frequencies:\n", observed_data)
print("\nExpected Frequencies:\n", expected.round(2))

```

```python
#16. Implement an F-test using Python to compare the variances of two random samples.

import numpy as np
from scipy.stats import f

def f_test(sample1, sample2, alpha=0.05):
    """
    Perform an F-test to compare the variances of two samples.

    Parameters:
        sample1 (array-like): First sample
        sample2 (array-like): Second sample
        alpha (float): Significance level (default = 0.05)

    Returns:
        dict: F-statistic, critical values, p-value, conclusion
    """
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)

    if var1 > var2:
        f_stat = var1 / var2
        dfn, dfd = len(sample1) - 1, len(sample2) - 1
    else:
        f_stat = var2 / var1
        dfn, dfd = len(sample2) - 1, len(sample1) - 1

    f_critical_low = f.ppf(alpha / 2, dfn, dfd)
    f_critical_high = f.ppf(1 - alpha / 2, dfn, dfd)

    p_value = 2 * min(f.cdf(f_stat, dfn, dfd), 1 - f.cdf(f_stat, dfn, dfd))

    conclusion = "Reject H₀ (variances are significantly different)" if p_value < alpha else "Fail to reject H₀ (variances are not significantly different)"

    return {
        "F-statistic": f_stat,
        "Degrees of freedom": (dfn, dfd),
        "P-value": p_value,
        "Critical values": (f_critical_low, f_critical_high),
        "Conclusion": conclusion
    }

    np.random.seed(0)

sample1 = np.random.normal(loc=50, scale=10, size=30)
sample2 = np.random.normal(loc=52, scale=15, size=30)

result = f_test(sample1, sample2)
for key, value in result.items():
    print(f"{key}: {value}")

```

```python
#19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro, levene
from statsmodels.stats.anova import anova_lm
import statsmodels.api as sm
from statsmodels.formula.api import ols

def check_anova_assumptions(data, group_col, value_col):
    """
    Check ANOVA assumptions: normality, homogeneity of variances, and visualize residuals.

    Parameters:
        data (DataFrame): pandas DataFrame containing the data
        group_col (str): name of the column with group labels
        value_col (str): name of the column with numeric values
    """
    # Fit the ANOVA model
    model = ols(f'{value_col} ~ C({group_col})', data=data).fit()
    residuals = model.resid

    print("1. Normality of Residuals (Shapiro-Wilk Test):")
    stat, p = shapiro(residuals)
    print(f"   W-statistic = {stat:.4f}, p-value = {p:.4f}")
    if p > 0.05:
        print("   ✔️ Residuals appear normal (fail to reject H₀)")
    else:
        print("   ❌ Residuals may not be normally distributed (reject H₀)")

    print("\n2. Homogeneity of Variance (Levene’s Test):")
    grouped_data = [group[value_col].values for name, group in data.groupby(group_col)]
    stat, p = levene(*grouped_data)
    print(f"   Levene-statistic = {stat:.4f}, p-value = {p:.4f}")
    if p > 0.05:
        print("   ✔️ Variances are equal across groups (fail to reject H₀)")
    else:
        print("   ❌ Variances may not be equal (reject H₀)")

    print("\n3. Independence of Observations:")
    print("   ✅ Usually ensured by study design (not testable with ANOVA directly)")

    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    sns.histplot(residuals, kde=True, color='skyblue')
    plt.title('Histogram of Residuals')

    plt.subplot(1, 2, 2)
    sm.qqplot(residuals, line='s', ax=plt.gca())
    plt.title('Q-Q Plot of Residuals')

    plt.tight_layout()
    plt.show()


import numpy as np
import pandas as pd

np.random.seed(42)
df = pd.DataFrame({
    'Group': ['A'] * 30 + ['B'] * 30 + ['C'] * 30,
    'Score': np.concatenate([
        np.random.normal(50, 5, 30),
        np.random.normal(52, 5, 30),
        np.random.normal(54, 8, 30)
    ])
})

check_anova_assumptions(df, group_col='Group', value_col='Score')
```

```python
#20. Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results

import numpy as np
import pandas as pd

np.random.seed(42)

# Create factors: Factor A (e.g., Treatment), Factor B (e.g., Gender)
factor_A = ['Low', 'High']
factor_B = ['Male', 'Female']

# Generate synthetic data
data = []
for a in factor_A:
    for b in factor_B:
        # Mean differs by interaction of A and B
        mean = 50 + (5 if a == 'High' else 0) + (3 if b == 'Female' else 0)
        scores = np.random.normal(loc=mean, scale=5, size=30)
        for score in scores:
            data.append({'Treatment': a, 'Gender': b, 'Score': score})

df = pd.DataFrame(data)

import statsmodels.api as sm
from statsmodels.formula.api import ols

# Fit the two-way ANOVA model with interaction
model = ols('Score ~ C(Treatment) * C(Gender)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)  # Type II ANOVA
print(anova_table)
```

```python
#23. Simulate random data from a normal distribution, then perform hypothesis testing to evaluate the means.

import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

# Step 1: Simulate data
np.random.seed(42)
population_mean = 50  # hypothesized population mean
sample = np.random.normal(loc=52, scale=5, size=30)  # sample from N(52, 5)

# Step 2: Perform one-sample t-test
t_stat, p_value = stats.ttest_1samp(sample, popmean=population_mean)

# Step 3: Display results
print("One-Sample T-Test")
print("==================")
print(f"Sample mean: {np.mean(sample):.2f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject H₀ — Sample mean is significantly different from population mean.")
else:
    print("Conclusion: Fail to reject H₀ — No significant difference from population mean.")

# Step 4: Visualize sample distribution
plt.figure(figsize=(8, 5))
plt.hist(sample, bins=10, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(np.mean(sample), color='red', linestyle='--', label='Sample Mean')
plt.axvline(population_mean, color='green', linestyle='--', label='Population Mean')
plt.title('Histogram of Simulated Sample')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
#24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt

# Step 1: Simulate sample data
np.random.seed(42)
sample = np.random.normal(loc=50, scale=6, size=25)  # True std = 6
sample_variance = np.var(sample, ddof=1)
n = len(sample)
sigma_0_sq = 36  # Hypothesized variance = 6²

# Step 2: Calculate Chi-square statistic
chi_sq_stat = (n - 1) * sample_variance / sigma_0_sq

# Step 3: Critical values and p-value (two-tailed)
alpha = 0.05
df = n - 1
chi2_lower = chi2.ppf(alpha / 2, df)
chi2_upper = chi2.ppf(1 - alpha / 2, df)
p_value = 2 * min(chi2.cdf(chi_sq_stat, df), 1 - chi2.cdf(chi_sq_stat, df))

# Step 4: Display results
print("Chi-Square Test for Population Variance")
print("=======================================")
print(f"Sample Variance: {sample_variance:.4f}")
print(f"Chi-Square Statistic: {chi_sq_stat:.4f}")
print(f"Degrees of Freedom: {df}")
print(f"Chi-Square Critical Values: [{chi2_lower:.4f}, {chi2_upper:.4f}]")
print(f"P-value: {p_value:.4f}")

if chi_sq_stat < chi2_lower or chi_sq_stat > chi2_upper:
    print("Conclusion: Reject H₀ — Sample variance differs significantly from population variance.")
else:
    print("Conclusion: Fail to reject H₀ — No significant difference from population variance.")

# Step 5: Visualize
x = np.linspace(chi2.ppf(0.001, df), chi2.ppf(0.999, df), 500)
y = chi2.pdf(x, df)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', label='Chi-square PDF')
plt.fill_between(x, y, where=(x < chi2_lower) | (x > chi2_upper), color='red', alpha=0.3, label='Rejection Region')
plt.axvline(chi_sq_stat, color='black', linestyle='--', label='Test Statistic')
plt.title('Chi-Square Distribution for Variance Test')
plt.xlabel('Chi-square Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
#26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Step 1: Simulate two datasets
np.random.seed(0)
data1 = np.random.normal(loc=50, scale=10, size=30)  # variance = 100
data2 = np.random.normal(loc=50, scale=5, size=30)   # variance = 25

# Step 2: Calculate variances and F-statistic
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

f_stat = var1 / var2
df1 = len(data1) - 1
df2 = len(data2) - 1

# Step 3: Compute p-value and critical values (two-tailed test)
alpha = 0.05
f_crit_low = f.ppf(alpha / 2, df1, df2)
f_crit_high = f.ppf(1 - alpha / 2, df1, df2)
p_value = 2 * min(f.cdf(f_stat, df1, df2), 1 - f.cdf(f_stat, df1, df2))

# Step 4: Display results
print("F-Test for Equality of Variances")
print("=================================")
print(f"Sample Variance 1: {var1:.4f}")
print(f"Sample Variance 2: {var2:.4f}")
print(f"F-statistic: {f_stat:.4f}")
print(f"Degrees of Freedom: ({df1}, {df2})")
print(f"P-value: {p_value:.4f}")
print(f"Critical Values: [{f_crit_low:.4f}, {f_crit_high:.4f}]")

if f_stat < f_crit_low or f_stat > f_crit_high:
    print("Conclusion: Reject H₀ — Variances are significantly different.")
else:
    print("Conclusion: Fail to reject H₀ — No significant difference in variances.")

# Step 5: Visualize F-distribution
x = np.linspace(0.1, 5, 500)
y = f.pdf(x, df1, df2)

plt.figure(figsize=(10, 6))
plt.plot(x, y, 'b-', label='F-distribution PDF')
plt.axvline(f_stat, color='black', linestyle='--', label=f'F-statistic = {f_stat:.2f}')
plt.fill_between(x, y, where=(x < f_crit_low) | (x > f_crit_high), color='red', alpha=0.3, label='Rejection Region')
plt.title('F-distribution with Rejection Regions')
plt.xlabel('F value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

```python
#27. Perform a Chi-square test for goodness of fit with simulated data and analyze the results.

import numpy as np
from scipy.stats import chisquare
import matplotlib.pyplot as plt

# Step 1: Simulate observed data (e.g., biased die rolls)
np.random.seed(42)
observed = np.random.choice([1, 2, 3, 4, 5, 6], size=120, p=[0.10, 0.10, 0.20, 0.20, 0.20, 0.20])
observed_counts = np.array([np.sum(observed == i) for i in range(1, 7)])

# Step 2: Define expected frequencies (for a fair die)
expected_counts = np.array([120 / 6] * 6)

# Step 3: Perform Chi-square test
chi_stat, p_value = chisquare(f_obs=observed_counts, f_exp=expected_counts)

# Step 4: Print the results
print("Chi-Square Goodness-of-Fit Test")
print("================================")
print(f"Observed counts: {observed_counts}")
print(f"Expected counts: {expected_counts}")
print(f"Chi-square Statistic: {chi_stat:.4f}")
print(f"P-value: {p_value:.4f}")

alpha = 0.05
if p_value < alpha:
    print("Conclusion: Reject H₀ — Observed data does not fit expected distribution.")
else:
    print("Conclusion: Fail to reject H₀ — No significant difference from expected distribution.")

# Step 5: Visualize
labels = ['1', '2', '3', '4', '5', '6']
x = np.arange(len(labels))

plt.figure(figsize=(8, 5))
plt.bar(x - 0.2, observed_counts, width=0.4, label='Observed', color='skyblue')
plt.bar(x + 0.2, expected_counts, width=0.4, label='Expected', color='lightgreen')
plt.xticks(x, labels)
plt.xlabel('Die Face')
plt.ylabel('Frequency')
plt.title('Observed vs Expected Frequencies')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

```
